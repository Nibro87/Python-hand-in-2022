```python
#Create a pie chart that shows the distribution of Licenses (PG, R, M and so on)
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv (r'../data/Highest Holywood Grossing Movies.csv')
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

group_by_diag = df.groupby("License").count().reset_index()
sizes = group_by_diag['Title']
labels = group_by_diag['License']

plt.pie(sizes, labels = labels, colors = colors)
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
#Calculate the average of world sales for each genre and visualize the data with a bar chart. (Hint: use groupBy)
df = pd.read_csv (r'../data/Highest Holywood Grossing Movies.csv')

average = df.groupby(['Genre'])['World Sales (in $)'].mean()

average

average.plot(kind= "bar", x='Genre')


```
